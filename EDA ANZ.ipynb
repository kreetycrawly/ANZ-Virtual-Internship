```python
import seaborn as sns
date_grp = df.groupby('date')["amount"].sum().reset_index() # total amounts of money transferred by day
plt.figure(figsize=(26, 6))
plt.xticks(rotation = 90)
sns.lineplot(y="amount",x="date",data=date_grp)
```

```python
date_mean = df.groupby("date")["amount"].mean().reset_index() # average amount of transactions per day
```

```python
plt.figure(figsize=(26, 6))
plt.xticks(rotation = 90)
sns.lineplot(y="amount",x="date",data=date_mean)
```

```python
g_v = df[["date","amount","gender"]]
g_d = g_v.groupby(['date',"gender"])["amount"].sum().reset_index()
plt.figure(figsize=(26, 6))
plt.xticks(rotation = 90)
sns.lineplot(y="amount",x="date",hue="gender",data=g_d)
# gender vs amount per day 
```

```python
g_m = df.groupby(["date","gender"])["amount"].mean().reset_index()
plt.figure(figsize=(26, 6))
plt.xticks(rotation = 90)
sns.lineplot(y="amount",x="date",hue="gender", data=g_m)
# average amount of transactions by gender each day
```

```python
merchant_state_total = df.groupby(["merchant_state","gender"])["amount"].sum().reset_index()
plt.figure(figsize=(16, 6))
sns.barplot(x="amount", y="merchant_state",  data=merchant_state_total)
# amount of transactions per state
```

```python
plt.figure(figsize=(16, 4))
sns.barplot(x="amount", y="merchant_state", hue="gender", data=merchant_state_total)
# amount of transactions by gender per state
```

```python
merchant_state_mean = df.groupby(["date","merchant_state"])["amount"].mean().reset_index()
plt.figure(figsize=(26, 7))
plt.xticks(rotation = 90)
sns.lineplot(y="amount",x="date",hue="merchant_state", data=merchant_state_mean)
# amount of transaction by state per day
```

```python
merchant_suburb= df.groupby(["merchant_suburb"])["amount"].sum().reset_index()
suburb_sorted=merchant_suburb.sort_values(by='amount', ascending=False)
print("Top 10 suburbs with highest amount of transactions over 91 days:")
print(suburb_sorted.head(10).values)
```
